```python
def tipo(col):
    
    mammals = ['Squirrel', 'Pig', 'Gorilla', 'Koala', 'Anteater', 'Bull', 'Mouse', 'Cat', 'Horse', 'Hamster',
        'Kangaroo', 'Wolf', 'Elephant', 'Sheep', 'Deer', 'Tiger', 'Cub', 'Dog', 'Bear', 'Hippo', 'Goat',
        'Rabbit', 'Lion', 'Monkey', 'Rhino', 'Cow']

    birds = ['Bird', 'Eagle', 'Penguin', 'Chicken', 'Duck', 'Ostrich']

    amphibians = ['Alligator', 'Frog', 'Octopus']

    if col in mammals:
        return "Mammal"
    elif col in birds:
        return "Bird"
    elif col in amphibians:
        return "Amphibian"
    else:
        pass
```

```python
df_personajes["Type"] = df_personajes["Type"].apply(tipo)
```

```python
def dondebixos(col):
    
    near_flowers = ['Flying near flowers', 'On flowers', 'On white flowers', 'Flying near blue/purple/black flowers']

    near_forest = ['On trees (any kind)', 'Shaking trees (hardwood or cedar only)', 'On hardwood/cedar trees',
    'On tree stumps', 'Disguised under trees', 'Shaking trees']

    near_ground = ['On the ground', 'From hitting rocks', 'On tree stumps','Pushing snowballs',
        'Disguised on shoreline', 'On flowers', 'Flying',
        'Underground (dig where noise is loudest)', 'On rocks/bushes','On beach rocks']

    near_capitalism = ['On rotten turnips', 'On villagers', 'Flying near light sources',
    'Flying near trash (boots, tires, cans) or rotten turnips']

    near_water = ['Flying near water', 'On palm trees', 'On rivers/ponds', 'Disguised on shoreline', 'On beach rocks']

    if col in near_flowers:
        return "Near flowers"
    elif col in near_forest:
        return "Near forest"
    elif col in near_ground:
        return "Near ground"
    elif col in near_capitalism:
        return "Near town"
    elif col in near_water:
        return "Near water"
```
